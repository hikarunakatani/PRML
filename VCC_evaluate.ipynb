```
#@title ←ここを押すとtest1が始まります　Press here to start test1
import numpy as np
import glob
import random
import re
import os
import IPython.display
import time
import itertools
from google.colab import files
def MOS():
    # 発話数
    setnum = 1
    # 評価音声の種類
    cvnum = 60
    dtype = [("file_name", "U50"), ("score", "i1")]
    score_list = np.zeros(setnum * cvnum, dtype=dtype)
    score_tmp = []
    filename_tmp = []
    rand = []
    file_list = []
    spk_list = ["TFF1", "TFM1", "TGF1", "TGM1", "TMF1", "TMM1"]
    methods = ["nat", "rec", "cyc", "int", "eng"]
    #print("あなたの名字を半角英字で入力してください。\n")
    #print("Input your first name here.\n")
    #while True:
    #    name = input()
    #    if re.match(r'[a-z]', name[:]) :
    #      break
    #    else:
    #        print("半角英字で入力してください。\n")
    print("音声を聴いて、その音声の音質（音が歪んでないかどうか）を5段階で評価してください（1:非常に悪い - 5:非常に良い）。\n")
    print("Evaluate the sound quality of voices (whether degraded/distorted or not) on a scale from 1 (very bad) to 5 (very good).\n")
    print("発音は無視してください。\n")
    print("Please ignore accent.\n")
    #time.sleep(3)
    for cv in methods:
          for spk_name in spk_list:
              if cv == "eng":
                file_list_tmp = glob.glob("./wav/MOS/" + str(spk_name) + "_" + str(cv) + "/eng/" + "*.wav")
                random.shuffle(file_list_tmp)
                file_list.append(file_list_tmp[:2])
                file_list_tmp.clear()
              else:
                file_list_tmp = glob.glob("./wav/MOS/" + str(spk_name) + "_" + str(cv) + "/SEF1/" + "*.wav")
                random.shuffle(file_list_tmp)
                file_list.append(file_list_tmp[:setnum])
                file_list_tmp.clear()
                file_list_tmp = glob.glob("./wav/MOS/" + str(spk_name) + "_" + str(cv) + "/SEM2/" + "*.wav")
                random.shuffle(file_list_tmp)
                file_list.append(file_list_tmp[:setnum])
                file_list_tmp.clear()
    file_list = itertools.chain(*file_list)
    file_list = list(file_list)
    random.shuffle(file_list)
    num = len(file_list)
    #print(file_list)
    for file_name in file_list:
        print("残り：%d文\n" %num)
        print("%d utterances left\n" %num)
        IPython.display.display(IPython.display.Audio(file_name))
        print("評点を記入してください(1:非常に悪い - 5:非常に良い)．\n")
        print("Input a score from 1 (very bad) to 5 (very good).")
        score = 0
        flag = True
        score = input()
        while flag == True:
            if re.match(r'[1-5]', score) and 1 <= int(score) <= 5:
                score = int(score)
                break
            else:
                print("1~5の数字を入力してください．")
                print("Input numbers from 1 to 5.")
                score = input()
        score_tmp.append(score)
        filename_tmp.append(file_name)
        num -= 1
    score_list["file_name"] = filename_tmp
    score_list["score"] = score_tmp
    np.savetxt("./csv/" + "VCC" + "_MOS.csv",
               score_list, fmt=["%s", "%d"])
    print("これでtest1は終わりです。test2に移ってください。\n")
    print("This is the end of test 1. Please move to test2.\n")
MOS()
```

```
#@title ←ここを押すとtest2が始まります　Press here to start test2
import numpy as np
import glob
import random
import re
import os
import itertools
import time
import IPython.display
from google.colab import files

def identity():
    # number of the voices used for evaluation
    setnum = 1
    # number of CV methods
    cvnum = 60
    dtype = [("file_name", "U50"), ("score", "i1")]
    file_list = []
    file_list_tmp = []
    score_list = np.zeros(setnum * cvnum, dtype=dtype)
    score_tmp = []
    filename_tmp = []
    rand = []
    spk_list = ["TFF1", "TFM1", "TGF1", "TGM1", "TMF1", "TMM1"]
    methods = ["nat", "rec", "cyc", "int", "eng"]
    flag=True
    #print("あなたの名字を半角英字で入力してください。\n")
    #print("Input your first name here.\n")
    print("これからテスト2を開始します。\n")
    print("The test2 will start soon.\n")
    #while True:
    #    name = input()
    #    if re.match(r'[a-z]', name[:]):
    #        break
    #    else:
    #        print("半角英字で入力してください。\n")

    print("音声が二つ出るので、声質が同じかかどうかを下の4項目から評価してください。\n")
    print("Please judge whether two voice samples could have been produced by the same speaker or not.\n")
    #time.sleep(1)
    print("1：同じ\n")
    print("2：多分同じ\n")
    print("3：多分違う\n")
    print("4：違う\n")
    print("1：Same\n")
    print("2：Same(not sure)\n")
    print("3：Different(not sure)\n")
    print("4：Different\n")
    #time.sleep(1)
    print("一部の音声は，多少音声が劣化/歪んでいるように聞こえる場合があります。しかし音声の歪みではなくあくまで声質に専念してください．\n")
    print("Some of the samples may sound somewhat degraded/distorted. Please try to listen beyond the distortion and concentrate on identifying the voice.\n")
    #time.sleep(1)
    for cv in methods:
          for spk_name in spk_list:
              if cv == "eng":
                file_list_tmp = glob.glob("./wav/MOS/" + str(spk_name) + "_" + str(cv) + "/eng/" + "*.wav")
                random.shuffle(file_list_tmp)
                file_list.append(file_list_tmp[:2])
                file_list_tmp.clear()
              else:
                file_list_tmp = glob.glob("./wav/MOS/" + str(spk_name) + "_" + str(cv) + "/SEF1/" + "*.wav")
                random.shuffle(file_list_tmp)
                file_list.append(file_list_tmp[:setnum])
                file_list_tmp.clear()
                file_list_tmp = glob.glob("./wav/MOS/" + str(spk_name) + "_" + str(cv) + "/SEM2/" + "*.wav")
                random.shuffle(file_list_tmp)
                file_list.append(file_list_tmp[:setnum])
                file_list_tmp.clear()
    #print(file_list)
    file_list = itertools.chain(*file_list)
    file_list = list(file_list)
    random.shuffle(file_list)
    num = len(file_list)
    for file_name in file_list:
        print("残り：%d文\n" %num)
        print("%d utterances left\n" %num)
        IPython.display.display(IPython.display.Audio(file_name))
        for spk_name in spk_list:
          if spk_name in file_name:
              file_list_tmp = glob.glob("./wav/MOS/" + spk_name + "_org/" + "*.wav")
              IPython.display.display(IPython.display.Audio(file_list_tmp[1]))
              file_list_tmp.clear()
        score = 0
        print("評点を記入してください。\n")
        print("1：同じ\n")
        print("2：多分同じ\n")
        print("3：多分違う\n")
        print("4：違う\n")
        print("Input a score from 1 to 4.\n")
        print("1：Same\n")
        print("2：Same(not sure)\n")
        print("3：Different(not sure)\n")
        print("4：Different\n")
        score = input()
        while flag == True:
            if re.match(r'[1-4]', score) and 1 <= int(score) <= 4:
                score = int(score)
                break
            else:
                print("1~4の数字を入力してください．")
                print("Input numbers from 1 to 4.")
                score = input()
        score_tmp.append(score)
        filename_tmp.append(file_name)
        num -= 1

    score_list["file_name"] = filename_tmp
    score_list["score"] = score_tmp
    #np.savetxt("./csv/" + str(name) + "_identity.csv",
    #           score_list, fmt=["%s", "%d"])
    np.savetxt("./csv/" + "VCC" + "_identity.csv",
               score_list, fmt=["%s", "%d"])
    print("これでtest2は終わりです。\n")
    print("This is the end of test2.")
identity()
```
